### Normalize GDB Data

```python
Broadcast = pd.read_csv(WorkingFolder + "Zone10_2009_01.gdb/output.csv", sep=",")
Broadcast.head(3)
```

```python
Broadcast.shape
```

```python
Vessel = pd.read_csv(WorkingFolder + "Zone10_2009_01.gdb/Vessel.csv", sep=",")
Vessel.head(3)
```

```python
Vessel.shape
```

```python
Voyage = pd.read_csv(WorkingFolder + "Zone10_2009_01.gdb/Voyage.csv", sep=",")
Voyage.head(3)
```

```python
Voyage.shape
```

```python
df = Vessel.merge(Voyage, left_on='MMSI', right_on='MMSI', how='inner')
df.head(3)
```

```python
df.shape
```

```python
df = df.merge(Broadcast, left_on=["MMSI", "VoyageID"], right_on=["MMSI", "VoyageID"], how="inner")
df.head()
```

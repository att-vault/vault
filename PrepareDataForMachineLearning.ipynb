### Prepare Data for Use Case: Classify Suspicious Activity from AIS Data

```python
# s3://vault-data-corpus/vessel data/ConsolidatedAIS/
WorkingFolder = "/Users/cv0361/Desktop/TechChallenge/Data/csv/ConsolidatedAIS/"

# s3://vault-data-corpus/vessel data/H2O Generated Data/
OutputDir = "/Users/cv0361/Desktop/TechChallenge/Data/csv/H2O Generated Data/"
```

```python
# Combining all Stats Data for All Years
df_list = list()

for FileName in glob(WorkingFolder + "Statistic*"):
    print(FileName)
    df = pd.read_csv(FileName, sep=",")
    
    print("Rows:", len(df))
    
    df_list.append(df)
    
#     break
    
Stat = pd.concat(df_list, ignore_index=True)
print("Total Rows:", Stat.shape)

# Remove duplicate vessel records after combining all the zones/years
Stat.drop_duplicates(inplace=True)

print("Non-Dup Total Rows:", Stat.shape)
# Vessel.reset_index(inplace=True)
```

```python
Stat.head()
```

```python
Vessel = pd.read_csv(WorkingFolder + "Vessel.csv", sep=",")

Vessel['vessel_type'] = Vessel['vessel_type'].fillna(0)
Vessel = Vessel.astype({"vessel_type": int})           # cast type to int

# Unified Vessel Type
Vessel.loc[Vessel.vessel_type.isin([30, 1001, 1002]), 'vessel_type'] = 1001                            # Fishing
Vessel.loc[Vessel.vessel_type.isin([36,37,1019]), 'vessel_type'] = 1019                                # Pleasure Craft/Sailing
Vessel.loc[Vessel.vessel_type.isin([1012,1013,1014,1015]), 'vessel_type'] = 1012                       # Passenger
Vessel.loc[Vessel.vessel_type.isin([31, 32, 52, 1025]), 'vessel_type'] = 1025                          # Tug Tow
Vessel.loc[Vessel.vessel_type.isin([80,81,82,83,84,85,86,87,88,89,1017, 1024]), 'vessel_type'] = 1024  # Tanker
Vessel.loc[Vessel.vessel_type.isin([70,71,72,73,74,75,76,77,78,79,1003,1004,1016]), 'vessel_type'] = 1016  # Cargo

Vessel.head()
```

```python
# Vessel.vessel_type.drop_duplicates().sort_values().tail(20)
```

```python
Vessel = Vessel.loc[Vessel.vessel_type.isin([1001,36,1012,1025,1024,1016])]
Vessel.shape
```

```python
df = Vessel.merge(Stat, left_on="mmsi_id", right_on="mmsi_id", how="inner")
df.head()
```

```python
df = df[['vessel_type', 'length', 'width', 'PingRecStart', 'PingRecEnd',
       'TotalPing', 'LatStd', 'LonStd', 'MaxSOG', 'MinSOG', 'MeanSOG',
       'MedianSOG', 'StdSOG', 'MaxCOG', 'MinCOG', 'MeanCOG', 'MedianCOG',
       'StdCOG', 'AnoThreshold', 'AnoClusterCount']]

df = df.drop_duplicates()
df.reset_index(inplace=True)
df.head()
```

```python
# Output Stat Data
df.to_csv(OutputDir + "AIS_Classify_VesselType.csv", index=None, header = True)
```
